```python
resource_path = 'C:/resources/EMG/02_27/'
classes = ['Rest', 'Eyebrow','Smile', 'Chew']

dataset = {}
for clas in classes:
    dataset[clas] = np.load(os.path.join(resource_path,clas+'.npy'))
    #datset[clas] = DCFilter(np.load(os.path.join(resource_path,clas+'.npy')))
```

```python
for clas in classes:

    print(clas)
    temp = []
    for i, data in enumerate(dataset[clas]):
        min, max, std = stat(data)
        print(std)
```

```python
std_threshold = 200
cut = 2000



for clas in classes:

    print(clas)
    temp = []
    for i, data in enumerate(dataset[clas]):
    
        start, end= getRange(data,std_threshold)
        sample = data[:,start:end]

        sample = signal.resample(sample, 100, axis = -1)

        sample = np.clip(sample, -cut, cut) 
        sample /= cut

        showMe(sample, [-1.1,1.1])
        if i == 5:
             break

    
```

```python
cutted = {}
for clas in classes:
 
    temp = []
    for i, data in enumerate(dataset[clas]):

        start, end= getRange(data,std_threshold)
        sample = data[:,start:end]

        sample = signal.resample(sample, 100, axis = -1)

        sample = np.clip(sample, -cut, cut) 
        sample /= cut
      
        temp.append(sample)

    cutted[clas] = np.array(temp)
    
           
```
